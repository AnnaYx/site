##Resposta questão 2



**a)**
$$ f(x) = x^4 - 4x^3 + 4x^2 $$

Primeiro, calculamos a derivada de $ f(x) $:
$$ f'(x) = 4x^3 - 12x^2 + 8x $$

Agora, igualamos $ f'(x) $ a zero para encontrar os pontos críticos:
$$ 4x^3 - 12x^2 + 8x = 0 $$
$$ x(4x^2 - 12x + 8) = 0 $$
$$ x(x^2 - 3x + 2) = 0 $$
$$ x(x-1)(x-2) = 0 $$

Portanto, os pontos críticos são $ x = 0 $ $ x = 1 $ e $ x = 2 $.

Para determinar a natureza desses pontos críticos, precisamos verificar o sinal da segunda derivada $ f''(x) $ nesses pontos:
$$ f''(x) = 12x^2 - 24x + 8 $$

Calculando $ f''(0) $ $ f''(1) $ e $ f''(2) $:
$$ f''(0) = 8 > 0 $$ (mínimo local)
$$ f''(1) = -4 < 0 $$ (máximo local)
$$ f''(2) = 8 > 0 $$ (mínimo local)

**b)**
$$ f(x) = x^4(x-3)^2 $$

Primeiro, calculamos a derivada de $ f(x) $
$$ f'(x) = 4x^3(x-3)^2 + x^4(2)(x-3)(1) $$
$$ f'(x) = 4x^3(x-3)^2 + 2x^4(x-3) $$

Igualamos $ f'(x) $ a zero para encontrar os pontos críticos:
$$ x^3(4(x-3)^2 + 2x(x-3)) = 0 $$
$$ x^3((4(x-3)+2x)(x-3)) = 0 $$
$$ x^3((4x-12+2x)(x-3)) = 0 $$
$$ x^3((6x-12)(x-3)) = 0 $$
$$ x^3(6(x-2)(x-3)) = 0 $$
$$ x^3(6)(x-2)(x-3) = 0 $$

Os pontos críticos são $ x = 0 $,$ x = 2 $, e $ x = 3 $.

# Questão 4 - Alternativa - Vale até 2 pontos

Em uma fábrica, latas de sopa cilíndricas devem ser fabricadas para conter um volume dado $V$. Não há desperdício de material na corte da parte vertical de cada lata ( o lado do cilíndro), mas cada tampa e fundo, que são círculos de raio $r$, são cortados de um quadrado que mede $2r$ unidades de lado. Assim, o material usado para fabricar cada lata de sopa tem uma área de $A = 2 \pi rh + 8r^2$ unidades quadradas. Encontre a proporção entre a altura e o diâmetro para a lata mais econômica (ou seja, que requer a menor quantidade de material para fabricação).


## Resposta questão 4 alternativa

A área de superfície $ A $ é dada por:

$$
A = 2\pi rh + 8r^2
$$

E o volume $ V $ do cilindro é:

$$
V = \pi r^2 h
$$

Primeiro, expressamos $ h $ em termos de $ r $ usando a fórmula do volume:

$$
h = \frac{V}{\pi r^2}
$$

Substituímos $ h $ na fórmula da área:

$$
A(r) = 2\pi r\left(\frac{V}{\pi r^2}\right) + 8r^2 = \frac{2V}{r} + 8r^2
$$

Agora, para encontrar o valor de $ r $ que minimiza $ A(r) $ derivamos $ A(r) $ em relação a $ r $ e igualamos a zero para encontrar os pontos críticos:

$$
A'(r) = -\frac{2V}{r^2} + 16r = 0
$$

Resolvendo para $ r $, obtemos:

$$
r^3 = \frac{V}{8}
$$

$$
r = \left(\frac{V}{8}\right)^{\frac{1}{3}}
$$

Agora encontramos $ h $:

$$
h = \frac{V}{\pi r^2} = \frac{V}{\pi\left(\left(\frac{V}{8}\right)^{\frac{1}{3}}\right)^2} = 2\left(\frac{V}{8}\right)^{\frac{1}{3}}
$$

A proporção entre a altura e o diâmetro é então:

$$
\frac{h}{2r} = \frac{2\left(\frac{V}{8}\right)^{\frac{1}{3}}}{2\left(\frac{V}{8}\right)^{\frac{1}{3}}} = 1
$$

Portanto, para a lata mais econômica, a altura deve ser igual ao diâmetro.

```python
import numpy as np
import matplotlib.pyplot as plt

# Função para calcular o diâmetro mais econômico em relação ao volume
def diametro_economico(V):
    r = (V / 8)**(1/3)  # Calcula o raio usando a fórmula obtida anteriormente
    return 2 * r  # Retorna o diâmetro (que é duas vezes o raio)

# Gerando valores para o volume (de 1 a 100 unidades cúbicas)
volumes = np.linspace(1, 100, 100)
diametros = [diametro_economico(V) for V in volumes]

# Criando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(volumes, diametros, label='Diâmetro Econômico')
plt.xlabel('Volume (unidades cúbicas)')
plt.ylabel('Diâmetro (unidades)')
plt.title('Relação entre Volume e Diâmetro para a Lata Mais Econômica')
plt.legend()
plt.grid(True)

# Mostrando o gráfico
plt.show()
```

#Questão 5 - Vale até 1 ponto

Seja a função $ F(x, y) = x^3 + 3x^2y - 2xy^2 + y^3 + \cos(x + 2y)$, encontre as suas derivadas parciais.

**Rubrica de Avaliação**: até um ponto se as derivadas forem corretamente encontradas.

##Resposta questão 5
$$ F(x, y) = x^3 + 3x^2y - 2xy^2 + y^3 + \cos(x + 2y) $$

1. Derivada parcial em relação a $ x $
   - Começamos aplicando a regra de derivação para cada termo da função em relação a $ x $ mantendo $ y $ como uma constante:

     $$
     \frac{\partial}{\partial x}(x^3) + \frac{\partial}{\partial x}(3x^2y) - \frac{\partial}{\partial x}(2xy^2) + \frac{\partial}{\partial x}(y^3) + \frac{\partial}{\partial x}(\cos(x + 2y))
     $$

   - Derivando termo a termo:
     - $\frac{\partial}{\partial x}(x^3) = 3x^2 $
     - $\frac{\partial}{\partial x}(3x^2y) = 6xy$ (derivada de $ x^2 $ é $ 2x $ multiplicada por $ 3y $)
     - $\frac{\partial}{\partial x}(2xy^2) = 2y^2 $ (derivada de $ x $ é $ 1 $ multiplicada por $ 2y^2 $
     - $\frac{\partial}{\partial x}(y^3) = 0$ (derivada de uma constante em relação a $ x $ é zero)
     - $\frac{\partial}{\partial x}(\cos(x + 2y)) = -\sin(x + 2y)$ (derivada do cosseno é menos o seno)
   - Reunindo os resultados:
     $$
     F_x(x, y) = 3x^2 + 6xy - 2y^2 - \sin(x + 2y)
     $$

2. **Derivada parcial em relação a $ y $**:
   - Seguimos o mesmo processo, mas agora derivando em relação a $ y $ e mantendo $ x $ constante:

     $$
     \frac{\partial}{\partial y}(x^3) + \frac{\partial}{\partial y}(3x^2y) - \frac{\partial}{\partial y}(2xy^2) + \frac{\partial}{\partial y}(y^3) + \frac{\partial}{\partial y}(\cos(x + 2y))
     $$
   - Derivando termo a termo:
     - $\frac{\partial}{\partial y}(x^3) = 0$ (derivada de uma constante em relação a $ y $ é zero)
     - $\frac{\partial}{\partial y}(3x^2y) = 3x^2 $ (derivada de $ y $ é $ 1 , multiplicada por $ 3x^2 $)
     - $\frac{\partial}{\partial y}(2xy^2) = -4xy$ (derivada de $ y^2 $ é $ 2y $, multiplicada por $ -2x $)
     - $\frac{\partial}{\partial y}(y^3) = 3y^2$
     - $\frac{\partial}{\partial y}(\cos(x + 2y)) = -2\sin(x + 2y)$
   - Reunindo os resultados:
     $$
     F_y(x, y) = 3x^2 - 4xy + 3y^2 - 2\sin(x + 2y)
     $$


```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função F(x, y)
def F(x, y):
    return x**3 + 3*x**2*y - 2*x*y**2 + y**3 + np.cos(x + 2*y)

# Definindo as derivadas parciais
def F_x(x, y):
    return 3*x**2 + 6*x*y - 2*y**2 - np.sin(x + 2*y)

def F_y(x, y):
    return 3*x**2 - 4*x*y + 3*y**2 - 2*np.sin(x + 2*y)

# Criando uma grade de valores para x e y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculando as derivadas parciais em cada ponto da grade
Zx = F_x(X, Y)
Zy = F_y(X, Y)

# Plotando o gráfico das derivadas parciais em relação a x
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.contourf(X, Y, Zx, cmap='viridis')
plt.colorbar()
plt.title('Derivada Parcial em relação a x')
plt.xlabel('x')
plt.ylabel('y')

# Plotando o gráfico das derivadas parciais em relação a y
plt.subplot(1, 2, 2)
plt.contourf(X, Y, Zy, cmap='viridis')
plt.colorbar()
plt.title('Derivada Parcial em relação a y')
plt.xlabel('x')
plt.ylabel('y')

plt.tight_layout()
plt.show()
```
